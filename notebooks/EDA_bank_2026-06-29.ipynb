# 2. Advanced Exploratory Data Analysis

The previous exploratory analysis provided a general overview of the dataset structure and the distribution of individual variables.

This section extends the EDA by exploring relationships between variables, identifying potential patterns associated with the target variable (`deposit`), and highlighting insights that may support subsequent stages such as Data Cleaning, Feature Engineering, Business Analysis and Predictive Modeling.

The objective of this analysis is exploratory only; no data transformations or cleaning operations are performed at this stage.

### Research Question 1

Do clients who subscribe to a term deposit have different account balances?

## 2.1 Relationship between Account Balance and Deposit Subscription

This analysis explores whether customers who subscribed to a term deposit show different account balance distributions compared with those who did not subscribe.

The objective is to identify preliminary financial patterns that may be relevant for later business analysis.

```python
# Create figure

plt.figure(figsize=(10,6))

# Boxplot

sns.boxplot(
    data=df,
    x="deposit",
    y="balance",
    palette="viridis"
)

plt.title("Account Balance by Deposit Subscription")

plt.xlabel("Deposit Subscription")

plt.ylabel("Average Yearly Balance (€)")

plt.show()
```

### Interpretation

The boxplot allows a preliminary comparison of account balance distributions between customers who subscribed to a term deposit and those who did not.

Particular attention should be paid to:

- Median balance
- Distribution spread
- Presence of extreme values
- Potential differences between both customer groups

At this stage, no causal conclusions are drawn. The analysis is intended to identify patterns that may justify further statistical investigation.

```python
df.groupby("deposit")["balance"].agg(
    ["count", "mean", "median", "std", "min", "max"]
).round(2)
```
